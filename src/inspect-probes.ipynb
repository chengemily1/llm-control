```python
import matplotlib.pyplot as plt
import torch
import numpy as np
import json
```

```python
results_format = '/home/echeng/llm-control/experiments/reviews/user_negative/probing_results/Meta-Llama-3-8B_layer_{}_rsNone_classification_results.json' # layer num
```

```python
layer = 0
```

```python
for layer in range(33):

    with open(results_format.format(layer), 'r') as f:
        results = json.load(f)
    
    fig, ax1 = plt.subplots()
    
    # Plot loss on the left y-axis
    ax1.plot(results['val_loss'], label='Loss', color='tab:blue')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Validation Loss', color='tab:blue')
    ax1.tick_params(axis='y')
    ax1.set_ylim([0, 1])
    ax1.grid()
    
    # Create a second y-axis for accuracy
    ax2 = ax1.twinx()
    ax2.plot(results['val_accuracy'], label='Accuracy', color='tab:orange')
    ax2.set_ylabel('Validation Accuracy', color='tab:orange', labelpad=14, rotation=270)
    ax2.tick_params(axis='y')
    
    plt.title(f'Validation Loss and Accuracy, Layer={layer}')
    plt.show()
```
